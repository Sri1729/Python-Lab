```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../Datasets/LungCancer.csv")
data
```

```python
data.info()
```

```python
# cancer yes
CY = data[data.LUNG_CANCER=="YES"]

# cancer no
CN = data[data.LUNG_CANCER=="NO"]

plt.title("Cancer yes vs no")
plt.xlabel("AGE")
plt.ylabel("SMOKING")
plt.scatter(CY.AGE, CY.SMOKING, color = "red", label = "Cancer Yes")
plt.scatter(CN.AGE, CN.SMOKING, color = "lime", label = "Cancer No")
plt.legend()
plt.show()
```

```python
data.LUNG_CANCER = [1 if i=="YES" else 0 for i in data.LUNG_CANCER]
data.GENDER = [0 if i=="M" else 1 for i in data.GENDER]
data
```

```python
x = data.drop(["LUNG_CANCER"], axis = 1)
y = data.LUNG_CANCER.values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
 print("Accuracy for the model is: {}%".format(round(nb.score(x_test, y_test)*100,4)))
```

```python
y_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Performance metrics for our modal is:")
print("Accuracy: {:.2f}%".format(accuracy*100))
print("Precision: {:.2f}%".format(precision*100))
print("Recall: {:.2f}%".format(recall*100))
```
